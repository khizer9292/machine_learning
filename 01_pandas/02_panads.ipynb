# Splitting the string into multiple columns 
spliting is mostly use in Time series

```python
df = pd.DataFrame({
    "Name":["Muhammad Adan", "Ahtesham Iqbal", "Danish Elihe", "Rustom Ali"],
    "Location": ["Lahore , Pakistan", "Omaha ,  USA", "Multan , Pakistan", "Lahore , Pakistan"],

})
df
```

```python
# spliting the name 
df.Name.str.split(" ", expand=True)
```

```python
# add this splits into two columns 
df[["first_name", "last_name"]]= df.Name.str.split(" ", expand=True)
df
```

```python
df[["city", "country"]] = df.Location.str.split(",",  expand=True)
df
```

```python
# Refine data manipulation
df = df[["first_name", "last_name", "city", "country"]]
df
```
