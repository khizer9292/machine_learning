# Feature Encoding
Feature encoding is process of converting `categorical features` into `numerical features`.
+ Label Encoding
+ Ordinal Encoding
+ One-Hot Encoding
+ Binary Encoding  

```python
# Import Libraies
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data Load
df = sns.load_dataset('tips')
df.head()
```

```python
df['time'].value_counts()
```

```python
# let's encode the time in labelencoder with sklearn
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
le = LabelEncoder()

df['encoded_time'] = le.fit_transform(df['time'])
df.sample(15)
```

```python
df['encoded_time'].value_counts()
```

# Ordinal Encoder 

```python
# first we check the value count of day column
df['day'].value_counts()
```

```python
# Now, we encode the value of day column by using Ordinal Encoder
oe = OrdinalEncoder(categories=[['Thur', 'Fri', 'Sat', 'Sun']])
df['encoded_day'] = oe.fit_transform(df[['day']])
df.sample(15)
```

```python
df['encoded_day'].value_counts(ascending=True)
```

# One-Hot Encoding

```python
# load the dataset from the seaborn library
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
# NOW, we use onehotencoder on the column named embarked in titanic dataset
# create the instance of OneHotEncoder class of sklearn library
onehot_encoder = OneHotEncoder(sparse_output=False)
embarked_ohe = onehot_encoder.fit_transform(titanic[['embarked']])
embarked_ohe_df = pd.DataFrame(embarked_ohe,  columns = onehot_encoder.get_feature_names_out(['embarked']))
embarked_ohe_df
```

```python
# we create intance of the onehotencoder and fit the data into it then tansform the data into a dataframe
onehot_encoder = OneHotEncoder(sparse_output=False)
embarked_ohe = onehot_encoder.fit_transform(titanic[['embarked']])
embarked_ohe_df = pd.DataFrame(embarked_ohe, columns=onehot_encoder.get_feature_names_out(['embarked']))
embarked_ohe_df
```

```python
# concatenate the dataframe with the orifginal dataframe
titanic = pd.concat([titanic.reset_index(drop=True), embarked_ohe_df.reset_index(drop=True)], axis=1)
titanic
```
