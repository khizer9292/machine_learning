# Inverse Encoding/ Inverse Transform/ Decoding categorical Variable
Inverse Encoding is the process of converting the encoded data back into the original format.This is done by using the `inverse_transform()` method. This method is available for the following Encoders:
- Label Encoder
- One Hot Encoder 
- Ordinal Encoder

```python
# import the necessary libraries
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder 
```

```python
# first we load the dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# we have to impute all the missing values in the dataset to encode the categorical features
# first we drop the dect column from the dataset
df.drop('deck',axis=1, inplace=True) 

# now we impute the missing values in the age , embarked and embark_town columns 
df['age'] = df['age'].fillna(df['age'].median())
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town']= df['embark_town'].fillna(df['embark_town'].mode()[0])
df.isnull().sum().sort_values(ascending=False)
```

```python
# we should use separat encoder of each categorical features, So when we inverse encode the same data we will get the original data
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
le_embarked_town = LabelEncoder()
le_who = LabelEncoder()
le_class = LabelEncoder()
le_adult_male = LabelEncoder()

df['sex'] = le_sex.fit_transform(df['sex'])
df['embarked'] = le_embarked.fit_transform(df['embarked'])
df['embark_town'] = le_embarked_town.fit_transform(df['embark_town'])
df['who'] = le_who.fit_transform(df['who'])
df['class'] = le_class.fit_transform(df['class'])
df['adult_male'] = le_adult_male.fit_transform(df['adult_male'])
```

```python
df.head()
```

```python
# if i want to reverse encode the same data
df['sex'] = le_sex.inverse_transform(df['sex'])
df['embarked'] = le_embarked.inverse_transform(df['embarked'])
df['embark_town']= le_embarked_town.inverse_transform(df['embark_town'])
df['who'] = le_who.inverse_transform(df['who'])
df['class'] = le_class.inverse_transform(df['class'])
df['adult_male'] = le_adult_male.inverse_transform(df['adult_male'])
```

```python
df.head()
```

# **One Hot Encoder:**

```python
# import the libraries which we need
from sklearn.preprocessing import OneHotEncoder

# load the dataset again 
df = sns.load_dataset('titanic')

cat_columns = ['sex' , 'embarked' , 'embark_town' , 'who' , 'class']

# create a onehotencoder object 
encoder = OneHotEncoder(sparse_output=False)
encoded_df = pd.DataFrame(encoder.fit_transform(df[cat_columns]))

# this line get the feature names from the cat_columns and add them to the dataframe of encode_df
encoded_df.columns = encoder.get_feature_names_out(cat_columns)
encoded_df.head()
```

```python
# concatenate the dataframes with the main dataframe named as df 
df = pd.concat([df, encoded_df], axis = 1)
df.drop(cat_columns, axis=1, inplace = True)
df.shape
df.head()
```

```python
# We need to extract the original categories for each encoded column
original_categories = {col: encoder.categories_[i] for i, col in enumerate(cat_columns)}
# Manual creation of feature names
feature_names = []
for i, col in enumerate(cat_columns):
    for category in encoder.categories_[i]:
        feature_names.append(f"{col}_{category}")

encoded_df = pd.DataFrame(encoded_df, columns=feature_names)

df.head()
```

```python
import requests
import pandas as pd

def get_abs_data(postal_code):
    url = f"https://api.abs.gov.au/data/{postal_code}"  # Example ABS API endpoint
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        return {
            "postal_code": postal_code,
            "population": data.get("population"),
            "median_income": data.get("median_income"),
            "median_age": data.get("median_age"),
        }
    else:
        print(f"Failed to fetch data for {postal_code}")
        return None

# Example usage
print(get_abs_data(3000))  # Replace with a valid postal code
```
