# Seaborn and Ploting

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```


## Important things for Plotting according to grammar of graphics

```python
# loading the dataset
df = sns.load_dataset("titanic")
```

```python
df.info()
```

```python
# creating scatter plot of age and fare with sex 
sns.scatterplot(data= df, x= "age", y= "fare", hue= "sex")
```

```python
# changing the size of the plot 
plt.figure(figsize=(16,6))
sns.scatterplot(data= df, x= "age", y= "fare", hue= "sex")
```

```python
# changing the colors of dots according to yourself 
# changing the size of the plot
plt.figure(figsize=(16, 6))

# creating custom color palette
palette = {'male' : "#4287f5" , "female" : "#f043d6"}
sns.scatterplot(data= df , x = 'age' , y = 'fare' , hue = 'sex' , palette = palette)

# labels of the plot
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age and Fare Scatterplot")
```

```python
# changing the colors of dots according to yourself 
# changing the size of the plot
plt.figure(figsize=(16, 6))

# creating custom color palette
palette = {'male' : "#4287f5" , "female" : "#f043d6"}
sns.scatterplot(data= df , x = 'age' , y = 'fare' , hue = 'sex' , palette = palette)

# labels of the plot
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age and Fare Scatterplot")

# saving the plots in png and svg format
plt.savefig("./outputs/plots/scatter_plot_fare_age_sex.png", dpi = 600)
plt.savefig("./outputs/plots/scatter_plot_fare_age_sex.svg", dpi = 600)
```

```python
plt.figure(figsize=(16, 6))
sns.set_theme(style ="ticks", font_scale=1.5,font="Times New Roman")
# creating custom color palette
palette = {'male' : "#4287f5" , "female" : "#f043d6"}
sns.scatterplot(data= df , x = 'age' , y = 'fare' , hue = 'sex' , palette = palette)

# labels of the plot
plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Age and Fare Scatterplot")
```

# Box plot

```python
df["class"].unique()
```

```python
# define the size of the plot
plt.figure(figsize=(16, 6))
sns.set_theme(style ="ticks", font_scale=1,font="Times New Roman")
# creating custom color palette
palette = {'First' : "#3ba30f" , "Second" : "#3be3c7", "Third":"#291db3"}
sns.boxplot(data= df , x = 'sex' , y = 'age' , hue = 'class' , palette = palette)

# labels of the plot
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Sex and Age Boxplot")
plt.legend(loc="upper right",title = "Class Name")
plt.xticks(rotation = 0)
```

```python
# define the size of the plot
plt.figure(figsize=(16, 6))
sns.set_theme(style ="ticks", font_scale=1,font="Times New Roman")
# creating custom color palette
palette = {'First' : "#3ba30f" , "Second" : "#3be3c7", "Third":"#291db3"}
sns.boxenplot(data= df , x = 'sex' , y = 'age' , hue = 'class' , palette = palette)

# labels of the plot
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Sex and Age Boxenplot")
plt.legend(loc="upper right",title = "Class Name")
plt.xticks(rotation = 0)
```

# Bar plot

```python
plt.figure(figsize=(16,6))
palette = {"First" : "#3ba30f" , "Second" : "#3be3c7", "Third":"#291db3"}
sns.barplot(data=df,x="sex", y = "age", hue="class", palette=palette, capsize=0.03)
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Sex and Age Barplot")
plt.legend(loc="upper right",title = "Class Name")
plt.xticks(rotation = 0)

```

```python
plt.figure(figsize=(16,6))
palette = {"First" : "#3ba30f" , "Second" : "#3be3c7", "Third":"#291db3"}
sns.violinplot(data=df,x="sex", y = "age", hue="class", palette=palette)
plt.xlabel("Sex")
plt.ylabel("Age")
plt.title("Sex and Age violinplot")
plt.legend(loc="upper right",title = "Class Name")
plt.xticks(rotation = 0)
```
